```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import os
import sys
from pathlib import Path
import xarray as xr
import cfgrib
import cartopy.crs as ccrs  # Projeções de mapas.
import cartopy.feature as cfeature  # Elementos geográficos.

#Import local libraries
import aux
```

```python
year='2018'
month='12'

base_dir='/pesq/dados/monan/users/madeleine.gacita/global_data/'
```

```python
era5_var="surface_solar_radiation_downwards"
era5_file_arg = base_dir+f"era5/{year}/{month}/surface_solar_radiation_downwards_Mhourly.nc"
path = Path(era5_file_arg)
ds = xr.open_dataset(path, engine="netcdf4")
```

```python
# Initialize variable to store ERA5 hourly mean data
ds_hourly_mean = None
era5_var_name = None
    
if not path.exists():
    print(f"File does not exist: {path!s}")
else:    
    # First try xarray+cfgrib (the common path)
    try:
        ds = try_cfgrib(str(path))
        print("Successfully opened with cfgrib/xarray")
        print("Dimensions:", ds.dims)
#         print("Time coordinate:", ds.time.values if 'time' in ds.coords else "Not found")
#         print("Step coordinate:", ds.step.values if 'step' in ds.coords else "Not found")
        
    except Exception as e:
        print(f"cfgrib/xarray open failed: {e}")
        import traceback
        traceback.print_exc()
        ds_hourly_mean = None
        era5_var_name = None
```

```python
if ds is not None:
##     Convert longitudes from 0-360 to -180 a 180
#     ds['longitude']=aux.lon_to_minus180_180(ds['longitude'].values)
    # Convert to instantaneous flux (W/m²)
    # SSRD is accumulated, so we need to difference within each day
    print("\nCalculating instantaneous flux from accumulated SSRD...")

    # Get step values in seconds for all steps
    all_step_seconds = np.array([pd.Timedelta(s).total_seconds() for s in ds['step'].values])
    
    # For the first step, use the accumulated value divided by the step time
    # For subsequent steps, use the difference divided by the time interval
    
    # Initialize flux array with same shape as original
    ssrd_flux_values = np.zeros_like(ds['ssrd'].values)
    
    # First step: accumulated value / step time
    ssrd_flux_values[:, 0, :, :] = ds['ssrd'].values[:, 0, :, :] / all_step_seconds[0]
    
    # Remaining steps: difference / time interval
    for i in range(1, len(all_step_seconds)):
        time_interval = all_step_seconds[i] - all_step_seconds[i-1]
        ssrd_flux_values[:, i, :, :] = (ds['ssrd'].values[:, i, :, :] - ds['ssrd'].values[:, i-1, :, :]) / time_interval
    
    # Create DataArray with proper coordinates
    ds['ssrd_flux'] = xr.DataArray(
        ssrd_flux_values,
        dims=ds['ssrd'].dims,
        coords=ds['ssrd'].coords,
        attrs={'long_name': 'Surface Solar Radiation Downwards Flux', 
               'units': 'W m-2',
               'description': 'Instantaneous flux calculated from accumulated SSRD'}
    )

    print(f"\nCreated ssrd_flux variable:")
    print(f"  Shape: {ds['ssrd_flux'].shape}")
    print(f"  Number of steps: {len(ds['ssrd_flux']['step'].values)}")
    print(f"  Step range: {ds['ssrd_flux']['step'].values[0]} to {ds['ssrd_flux']['step'].values[-1]}")

    # Verify no NaN values from the calculation
    nan_count = np.isnan(ds['ssrd_flux'].values).sum()
    print(f"  NaN values: {nan_count} ({100*nan_count/ds['ssrd_flux'].size:.2f}%)")
    
    # Print sample values for verification
    print(f"\n  Sample flux values at first time, mid-grid point:")
    mid_lat = ds['ssrd_flux'].shape[2] // 2
    mid_lon = ds['ssrd_flux'].shape[3] // 2
    for i in range(min(5, len(ds['step'].values))):
        step_hours = all_step_seconds[i] / 3600
        flux_val = ds['ssrd_flux'].values[0, i, mid_lat, mid_lon]
        accum_val = ds['ssrd'].values[0, i, mid_lat, mid_lon]
        print(f"    Step {int(step_hours):02d}h: flux={flux_val:.2f} W/m², accumulated={accum_val:.2f} J/m²")

    print("\nDataset with ssrd_flux:")
    print(ds)
    
    # Save the step means to a NetCDF file
    output_file = f"{path.parent}/monthly_mean_by_step_{era5_var}_{year}{month}.nc"
    print(f"\nSaving monthly means by step to: {output_file}")

    # Use netcdf4 engine to avoid potential import issues
    ds['ssrd_flux'].to_netcdf(output_file, engine='netcdf4')
    print(f"Successfully saved to: {output_file}")
```

```python
# Create actual datetime by combining time and step
# time is the forecast initialization time, step is the forecast hour
if 'time' in ds.coords and 'step' in ds.coords:
    print("Available steps:", ds.step.values)
    print("Number of time values:", len(ds.time.values))    
    try:
        # Get the data variable name before any transformations
        era5_var_name = list(ds.data_vars)[0]
        print(f"Processing variable: {era5_var_name}")
        # Calculate mean across all time values for each step
        print("Calculating mean for each step (forecast hour)...")
        with xr.set_options(use_flox=False):
            ds_hourly_mean['ssrd_flux'] = ds['ssrd_flux'].mean(dim='time')
        print("\nMean calculated for each step:")
        print(ds_hourly_mean.dims)
        print("Steps:", ds_hourly_mean.step.values)       
        # Convert longitude from 0-360 to -180 to 180 format if needed
        if 'longitude' in ds_hourly_mean.coords:
            lon_values = ds_hourly_mean['longitude'].values
            if lon_values.max() > 180:
                print("Converting longitude from 0-360° to -180-180° format...")
                # Convert longitude: values > 180 become negative
                lon_converted = np.where(lon_values > 180, lon_values - 360, lon_values)
                ds_hourly_mean['longitude'] = lon_converted
                # Sort by longitude to maintain proper ordering
                ds_hourly_mean = ds_hourly_mean.sortby('longitude')
                print("Longitude converted to East-West format")
                
        # Save the step means to a NetCDF file
        output_file = f"{path.parent}/monthly_mean_by_step_{era5_var}_{year}{month}.nc"
        print(f"\nSaving monthly means by step to: {output_file}")

        # Use netcdf4 engine to avoid potential import issues
        ds_hourly_mean.to_netcdf(output_file, engine='netcdf4')
        print(f"Successfully saved to: {output_file}")
    except Exception as process_error:
        print(f"Error during processing: {process_error}")
        import traceback
        traceback.print_exc()
        raise
```

```python
if 'time' in ds.coords and 'step' in ds.coords:
    # Stack time and step into a single dimension
    ds_stacked = ds.stack(forecast_time=('time', 'step'))

    # Calculate actual time for each forecast (time + step)
    import pandas as pd
    import numpy as np

    # Ensure time is datetime64 and step is timedelta64
    time_values = ds_stacked.time.values
    step_values = ds_stacked.step.values

    # Convert to proper types if needed
    if not np.issubdtype(time_values.dtype, np.datetime64):
        time_values = time_values.astype('datetime64[ns]')
    if not np.issubdtype(step_values.dtype, np.timedelta64):
        step_values = step_values.astype('timedelta64[ns]')

    # Add them to get actual time
    actual_time_values = time_values + step_values

    # Ensure result is datetime64
    actual_time_values = actual_time_values.astype('datetime64[ns]')

    print("Debug - time sample:", time_values[:3])
    print("Debug - step sample:", step_values[:3])
    print("Debug - actual_time sample:", actual_time_values[:3])
    print("Debug - actual_time dtype:", actual_time_values.dtype)

    try:
        # Get the data variable name before any transformations
        era5_var_name = list(ds.data_vars)[0]
        print(f"Processing variable: {era5_var_name}")

        # Reset the MultiIndex to convert it to regular coordinates
        print("Resetting index...")
        ds_stacked = ds_stacked.reset_index('forecast_time')

        # Drop time and step (forecast_time is automatically handled by reset_index)
        print("Dropping old coordinates...")
        vars_to_drop = []
        if 'time' in ds_stacked.coords:
            vars_to_drop.append('time')
        if 'step' in ds_stacked.coords:
            vars_to_drop.append('step')
        if vars_to_drop:
            ds_stacked = ds_stacked.drop_vars(vars_to_drop)

        # Rename the forecast_time dimension to actual_time
        print("Renaming dimension...")
        ds_stacked = ds_stacked.rename({'forecast_time': 'actual_time'})

        # Assign the actual datetime values to the actual_time coordinate
        print("Assigning datetime coordinate...")
        ds_stacked = ds_stacked.assign_coords(actual_time=actual_time_values)

        print("Created actual_time coordinate")
        print("Sample times:", ds_stacked.actual_time.values[:5])
        print("Coordinate dtype:", ds_stacked.actual_time.dtype)

        # Extract hour of day from actual_time
        # Create hour as a regular numpy array to avoid dt accessor issues
        print("Extracting hours...")
        hours = pd.DatetimeIndex(actual_time_values).hour

        # Add hour as a data variable first, then set it as a coordinate
        ds_stacked['hour'] = ('actual_time', hours)
        ds_stacked = ds_stacked.set_coords('hour')

        # Group by hour and calculate mean across all times
        # Disable flox to avoid TypeAlias import error
        print("Grouping by hour and calculating means...")
        with xr.set_options(use_flox=False):
            ds_hourly_mean = ds_stacked.groupby('hour').mean(dim='actual_time')

        print("\nMonthly mean at each hour:")
        print(ds_hourly_mean.dims)
                    
        # Convert longitude from 0-360 to -180 to 180 format if needed
        if 'longitude' in ds_hourly_mean.coords:
            lon_values = ds_hourly_mean['longitude'].values
            if lon_values.max() > 180:
                print("Converting longitude from 0-360° to -180-180° format...")
                # Convert longitude: values > 180 become negative
                lon_converted = np.where(lon_values > 180, lon_values - 360, lon_values)
                ds_hourly_mean['longitude'] = lon_converted
                # Sort by longitude to maintain proper ordering
                ds_hourly_mean = ds_hourly_mean.sortby('longitude')
                print("Longitude converted to East-West format")
                    

        # Save the hourly means to a NetCDF file
        output_file = f"{base_dir}/era5//{year}/{month}/monthly_mean_hourly_{era5_var}.nc"
        print(f"\nSaving monthly hourly means to: {output_file}")

        # Use netcdf4 engine to avoid potential import issues
        ds_hourly_mean.to_netcdf(output_file, engine='netcdf4')
        print(f"Successfully saved to: {output_file}")
                    
        # Optional: Print statistics for each hour
        print(f"\nHourly mean statistics for {era5_var_name}:")
        for hour in range(24):
            if hour in ds_hourly_mean['hour'].values:
                hour_data = ds_hourly_mean[era5_var_name].sel(hour=hour)
                print(f"  Hour {hour:02d}: mean={float(hour_data.mean()):.4f}, "
                      f"min={float(hour_data.min()):.4f}, max={float(hour_data.max()):.4f}")
    except Exception as process_error:
        print(f"Error during processing: {process_error}")
        import traceback
        traceback.print_exc()
        raise
else:
    print("Could not find 'time' and 'step' coordinates to create datetime")
    ds_hourly_mean = None
    era5_var_name = None



print(ds_hourly_mean)
```
