```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import os
import sys
from pathlib import Path
import xarray as xr
import cfgrib
import cartopy.crs as ccrs  # Projeções de mapas.
import cartopy.feature as cfeature  # Elementos geográficos.

#Import local libraries
import aux
```

```python
year='2018'
month='12'

base_dir='/pesq/dados/monan/users/madeleine.gacita/global_data/'
var_name='ssrd'
var_flux_name=f'{var_name}_flux'
print(var_flux_name)
```

```python
era5_var="surface_solar_radiation_downwards"
era5_file_arg = base_dir+f"era5/{year}/{month}/surface_solar_radiation_downwards_flux_MHour.nc"
path = Path(era5_file_arg)
ds_era5 = xr.open_dataset(path, engine="netcdf4")
```

```python
fig, ax = plt.subplots(figsize=(10, 8), subplot_kw=dict(projection=ccrs.PlateCarree()))

# Plot the data (sorted by latitude to avoid ValueError)
im = ds_era5[var_flux_name].sel(hour=18).sortby('latitude').plot(
    ax=ax, 
    cmap='viridis', 
    vmax=1400,
    cbar_kwargs={
        'shrink': 0.5,  # Make colorbar 70% of the plot height
        'aspect': 25,   # Control width (higher = thinner)
        'pad': 0.05,    # Space between plot and colorbar
        'label': 'ERA5 Incoming Surface Solar Radiation (W/m²)'
    }
)

# Add geographic features
ax.coastlines(linewidth=0.5)
ax.add_feature(cfeature.BORDERS, linewidth=0.3)
ax.add_feature(cfeature.STATES, linewidth=0.2)

# Add gridlines with labels
gl = ax.gridlines(draw_labels=True, linewidth=0.5, color='gray', alpha=0.5, linestyle='--')
gl.top_labels = False
gl.right_labels = False
gl.xlabel_style = {'size': 10}
gl.ylabel_style = {'size': 10}

# Set title
ax.set_title(f'ERA5 {var_flux_name} at 18 UTC - {year}-{month}', fontsize=12, pad=20)

plt.tight_layout()
```

```python
target_lon = -60.0
# Plot ERA5 Mhourly mean data at 15 UTC and -45° longitude
if ds_era5 is not None and var_flux_name is not None:
    try:
        print(f"\nPlotting ERA5 data at 15 UTC, {target_lon}° longitude...")

        # Get the data variable at hour 15
        era5_data_2d = ds_era5[var_flux_name].sel(hour=15)

        # Find longitude index closest to target_lon
        era5_lon = era5_data_2d['longitude'].values
        era5_lon_idx = np.argmin(np.abs(era5_lon - target_lon))
        actual_lon = era5_lon[era5_lon_idx]

        # Extract slice at that longitude
        era5_slice = era5_data_2d.isel(longitude=era5_lon_idx)
        era5_lat = era5_slice['latitude'].values
        era5_values = era5_slice.values

        # Create the plot
        plt.figure(figsize=(8, 6))
        plt.plot(era5_values, era5_lat, label=f"ERA5 {var_flux_name}", linewidth=2)
        plt.xlabel('ERA5 Incoming Surface Solar Radiation (W/m²)')
        plt.ylabel('Latitude (°)')
        plt.title(f"ERA5 {era5_var} - Monthly Mean at 15 UTC, lon={actual_lon:.2f}°\n{year}-{month}")
        plt.grid(True, alpha=0.3)
        plt.legend()

        # Save the plot
        out_png = f"ERA5_{era5_var}_{year}{month}_15UTC_lon{int(actual_lon)}.png"
        plt.savefig(out_png, dpi=150, bbox_inches='tight')
        print(f"Saved ERA5 plot to: {out_png}")
        plt.close()

    except Exception as plot_error:
        print(f"Could not plot ERA5 data: {plot_error}")
        import traceback
        traceback.print_exc()
else:
    print("ERA5 hourly mean data not available for plotting")    
```
